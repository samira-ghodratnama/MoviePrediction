## Dealing with categorical data

```python
# Create Dictionaries with key as category name and value with a count of categories
## i.e frequency of that category in each categorical column.
creative_type_dict = training_df['creative_type'].value_counts()
source_dict = training_df['source'].value_counts()
production_method_dict = training_df['production_method'].value_counts()
genre_dict = training_df['genre'].value_counts()
language_dict = training_df['language'].value_counts()
board_rating_dict = training_df['board_rating_reason'].value_counts()
movie_board_display_dict = training_df['movie_board_rating_display_name'].value_counts()
movie_release_pattern_dict = training_df['movie_release_pattern_display_name'].value_counts()
```

```python
# Create a new column which acts as a weight 
# for that category and map with its respective dictionary.
training_df['encoded_creative_type'] = training_df['creative_type'].map(creative_type_dict)
training_df['encoded_source'] = training_df['source'].map(source_dict)
training_df['encoded_production_method'] = training_df['production_method'].map(production_method_dict)
training_df['encoded_genre'] = training_df['genre'].map(genre_dict)
training_df['encoded_language'] = training_df['language'].map(language_dict)
training_df['encoded_board_rating'] = training_df['board_rating_reason'].map(board_rating_dict)
training_df['encoded_movie_board_display'] = training_df['movie_board_rating_display_name'].map(movie_board_display_dict)
training_df['encoded_movie_release_pattern'] = training_df['movie_release_pattern_display_name'].map(movie_release_pattern_dict)
```

```python
# Drop Orginal Columns
training_df.drop(['creative_type', 'source', 'production_method', 'genre', 'language', 'board_rating_reason',
                 'movie_board_rating_display_name', 'movie_release_pattern_display_name'], axis = 1)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, hamming_loss

ModelsPerformance = {}
def metricsReport(modelName, test_labels, predictions):
    accuracy = accuracy_score(test_labels, predictions)

    macro_precision = precision_score(test_labels, predictions, average='macro')
    macro_recall = recall_score(test_labels, predictions, average='macro')
    macro_f1 = f1_score(test_labels, predictions, average='macro')

    micro_precision = precision_score(test_labels, predictions, average='micro')
    micro_recall = recall_score(test_labels, predictions, average='micro')
    micro_f1 = f1_score(test_labels, predictions, average='micro')
#    // hamLoss = hamming_loss(test_labels, predictions)
    print("------" + modelName + " Model Metrics-----")
    print("Accuracy: {:.4f}\nPrecision:\n  - Macro: {:.4f}\n  - Micro: {:.4f}\nRecall:\n  - Macro: {:.4f}\n  - Micro: {:.4f}\nF1-measure:\n  - Macro: {:.4f}\n  - Micro: {:.4f}"\
          .format(accuracy, macro_precision, micro_precision, macro_recall, micro_recall, macro_f1, micro_f1))
    ModelsPerformance[modelName] = micro_f1
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.multiclass import OneVsRestClassifier

bagClassifier = OneVsRestClassifier(BaggingClassifier(n_jobs=-1))
bagClassifier.fit(X_train, y_train)
bagPreds = bagClassifier.predict(X_test)
metricsReport("Bagging", y_test, bagPreds)

# bagPreds = bagClassifier.predict(X_test)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
from numpy import absolute
import warnings
warnings.filterwarnings('ignore')

# define model
xgbmodel = XGBRegressor(objective = 'reg:squarederror',
                                  eta = 0.1,
                                  max_depth = 3,
                                  min_child_weight = 3,
                                  subsample = 0.8,
                                  gamma = 1.45,
                                  colsample_bytree = 0.7,
                                  eval_metric = 'rmse',
                                  seed = 42,
                                  n_estimators = 1000)
# xgbmodel = XGBRegressor(n_estimators=1000, max_depth=7, eta=0.1, subsample=0.7, colsample_bytree=0.8)
# define model evaluation method
cv = RepeatedKFold(n_splits=5, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(xgbmodel, X_train, Y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (scores.mean(), scores.std()))
```
